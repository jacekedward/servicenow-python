## init data

```python
# --- ServiceNow instance + credentials ---
# In the left sidebar, click the ðŸ”‘ key icon, This opens Secrets / User data, change/add there
from google.colab import userdata
instance = userdata.get('instance')
user = userdata.get('user')
pwd = userdata.get('password')
```

## querying ServiceNow

```python
import requests
from requests.auth import HTTPBasicAuth
import pandas as pd
# variables instance, user, pwd are defined in "init data" section above
# --- Endpoint for Table API (example: incident table) ---
url = f"https://{instance}.service-now.com/api/now/table/u_iam_role_entitlement_concepts"

params = {
    'sysparm_limit': 8,
    'sysparm_display_value': 'true',#,
   # 'sysparm_query': 'name=glide.ui.report.old_report_builder^ORDERBYDESCsys_created_on'#,
    'sysparm_query': 'ORDERBYDESCsys_updated_on',
   'sysparm_fields': 'u_iam_reviewed, u_application_system,sys_updated_on,u_description,u_valid_to,u_status,sys_updated_by'   # <-- only return these two fields
   #'sysparm_query': 'ORDERBYDESCsys_created_on'
}

df = pd.DataFrame()

# --- Call API ---
response = requests.get(url, auth=HTTPBasicAuth(user, pwd), params=params)

# --- Check response ---
if response.status_code == 200:
    data = response.json()
    records = data['result']

    # Convert to DataFrame for easy use in Colab
    df = pd.DataFrame(records)
    from google.colab import data_table
    data_table.DataTable(df, include_index=False)
else:
    print("Error:", response.status_code, response.text)

import pandas as pd

df['sys_updated_on_berlin'] = (
    pd.to_datetime(df['sys_updated_on'], errors='coerce')
      .dt.tz_localize('America/Los_Angeles', ambiguous='infer', nonexistent='shift_forward')
      .dt.tz_convert('Europe/Berlin')
)

df
```
